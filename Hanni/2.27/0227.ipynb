问题：可以实例化，可以调用load(); 简单来说就是跟别的官方给的loader的调用规范一致  

理解：也就是说封装成类  

官方用法例子1：  
loader = TextLoader(file_path)  
docs = loader.load()  
官方用法例子2：  
loader = PyPDFLoader(file_path)  
docs = loader.load  

```python
import os
from langchain_community.document_loaders import TextLoader, JSONLoader, UnstructuredExcelLoader, UnstructuredHTMLLoader, PyPDFLoader
import random

# 封装成类    
class MultiFormatFileLoader:

    def __init__(self, file_path):
        self.file_path = file_path

    def load(self):
        # 获取文件类型
        _, file_type = os.path.splitext(self.file_path)
        file_type = file_type.lower()
        # 不同文件类型传入不同处理函数
        if file_type == ".md" or file_type == ".txt" or file_type == ".csv":
            loader = TextLoader(self.file_path)
        elif file_type == ".json":
            loader = JSONLoader(file_path=self.file_path,jq_schema=".city")
        elif file_type == ".xlsx":
            loader = UnstructuredExcelLoader(self.file_path, mode="elements")
        elif file_type == ".pdf":
            loader = PyPDFLoader(self.file_path)
        elif file_type == ".html":
            loader = UnstructuredHTMLLoader(self.file_path)
        else:
            raise ValueError('Unsupported file format')
        return loader.load()
        
        
# 测试
type_list = ['.md', '.txt', '.csv', '.json', '.xlsx', '.pdf', '.html']
random_choice = random.choice(type_list)
print('测试用例', random_choice, '类型')
MultiFormat_file="./test_load/example"+random_choice

# 实例化
loader = MultiFormatFileLoader(MultiFormat_file)
docs = loader.load()
print(docs[0].page_content) 
print(docs[0].metadata)
```

### RAG 
RAG 检索增强生成 Retrieval Augmented Generation  

通过检索获取相关的知识并将其融入Prompt，让大模型能够参考相应的知识从而给出合理回答。  

完整的RAG应用流程主要包含两个阶段： 
数据准备阶段：数据提取——>文本分割——>向量化（embedding）——>数据入库  
应用阶段：用户提问——>数据检索（召回）——>注入Prompt——>LLM生成答案  

文本、分词、向量化，word to vec的过程
-> 写入数据库
-> 数据相似性检索：计算查询向量与所有存储向量的相似性得分(余弦相似性、欧氏距离、曼哈顿距离等)

```python
from langchain_ollama import OllamaEmbeddings
from langchain.vectorstores import Chroma
from langchain.schema import Document

embedding_model = OllamaEmbeddings(
    base_url="http://localhost:11434", 
    model="deepseek-r1:1.5b"  
)

texts = [
    "机器学习是研究通过经验自动改进算法的学科。",
    "深度学习是机器学习的一个子集，处理神经网络。",
    "自然语言处理是人工智能的一个领域，关注计算机与人类的语言互动。",
    "人工智能正在改变世界。"
]

# 嵌入向量化
embeddings = embedding_model.embed_documents(texts)
print(len(embeddings))
for i in range(len(embeddings)):
    print(f"Embedding {i}: {embeddings[i]}")
```

```python
# 存储 使用文档和嵌入模型生成向量数据库
documents = [Document(page_content=text) for text in texts]
vectorstore = Chroma.from_documents(
    documents=documents, 
    embedding=embedding_model  # 使用Ollama模型生成嵌入向量
)

# 检索
query_text = "人工智能如何改变生活？"
# 在向量数据库中搜索最相似的文档片段
similar_documents = vectorstore.similarity_search(query_text)

# 最相似的文档片段
print(similar_documents[0])  
```
